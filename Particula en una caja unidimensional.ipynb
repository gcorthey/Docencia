```python
import numpy as np 
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,1,0.01)
a = np.max(x)
```

```python
#Cálculo y graficación de la función de onda (psi)
def fi(n,x,a):
    return 2*(np.sin((np.pi)*n/a*x))    

total_n =4
plt.rcParams["figure.figsize"] = [5,9]
f, axarr = plt.subplots(total_n, sharex=True)
for n in range(total_n):
        axarr[total_n-n-1].plot(x/a,fi(n+1,x,a))
        axarr[total_n-n-1].set_title('n ='+ str(n+1))
        axarr[total_n-n-1].plot(x/a,np.zeros(len(x)),'grey')
        axarr[total_n-n-1].set_yticks([])
        axarr[total_n-n-1].set_xlim((0, 1))  
```

```python
#Cálculo y graficación del complejo conjugado de la función de onda (psi psi*)
def fi_2(n,x,a):
    return 4*(np.sin((np.pi)*n/a*x))**2    
total_n =4
plt.rcParams["figure.figsize"] = [5,9]
f, axarr = plt.subplots(total_n, sharex=True)
for n in range(total_n):
        axarr[total_n-n-1].plot(x/a,fi_2(n+1,x,a))
        axarr[total_n-n-1].set_title('n ='+ str(n+1))
        axarr[total_n-n-1].plot(x/a,np.zeros(len(x)),'grey')
        axarr[total_n-n-1].set_yticks([])
        axarr[total_n-n-1].set_xlim((0, 1))  
```
